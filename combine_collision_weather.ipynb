```python
import pandas as pd
```

```python
file_path = "Result/Traffic_Collision_Data_2016.csv"
collision_2016 = pd.DataFrame(pd.read_csv(file_path))
collision_2016.head()
```

```python
collision_2016_groupbyday = collision_2016.groupby("Date Occurred")
accident_per_day = pd.DataFrame(collision_2016_groupbyday["Area ID"].count()).reset_index()
accident_per_day=accident_per_day.rename(columns={"Area ID" : "Collision Count"})

accident_per_day.head()
```

```python
weather_path = "Resources/weather_date.csv"
weather_2016 = pd.DataFrame(pd.read_csv(weather_path))
weather_2016.head()
```

```python
daily_2016 = weather_2016
daily_2016["Collision Count"]=accident_per_day["Collision Count"]
daily_2016.to_csv("Result/Daily_weather_and_collision_2016.csv",index=False)
daily_2016.head()
```
